```python
x = df.iloc[:, :-1].values
x
```

```python
y= df.iloc[:,-1].values
y
```

```python
#splitting into training and splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#normalisation
#standardscalar
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train_nor=sc.transform(x_train)
x_test_nor=sc.transform(x_test)
x_train_nor
```

```python
x_test_nor
```

```python
# Import the classifier
from sklearn.neighbors import KNeighborsClassifier

# Create the model (e.g., with k=7)
kml = KNeighborsClassifier(n_neighbors=7)

# Fit the model (requires both x_train and y_train)
kml.fit(x_train_nor, y_train)

# Predict using the test set
y_pred = kml.predict(x_test_nor)

# Output predictions
print(y_pred)
```

```python
y_test
```

```python
#performance evaluation
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
